```python
import pandas as pd
import matplotlib.pyplot as plt

#import the data:
jobs = pd.read_csv('job-automation-probability.csv')
jobs.columns
```

```python


# Histogram of the variable 'probability'
plt.hist(jobs['probability'], bins=20)
plt.title('Histogram of Probability')
plt.xlabel('Probability')
plt.ylabel('Frequency')
plt.show()

```

```python
# Scatter plot of 'Average annual wage' and 'numbEmployed'
plt.scatter(jobs['Average annual wage'], jobs['numbEmployed'])
plt.title('Scatter Plot of Average Annual Wage and Number of Employed')
plt.xlabel('Average Annual Wage')
plt.ylabel('Number of Employed')
plt.show()


```
